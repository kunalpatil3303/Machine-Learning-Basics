```python
import easyocr as ocr
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/LP_USA_California_passenger.jpg")
```

```python
plt.imshow(img)
```

```python
reader = ocr.Reader(['en'])
```

```python
results = reader.readtext(img)
```

```python
print(results)
```
