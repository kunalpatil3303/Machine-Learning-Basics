Textual data pre-processing

```python
import numpy as np
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
```

```python
nltk.download('stopwords')
```

```python
print(stopwords.words('english'))
```

```python
news_data = pd.read_csv("/content/fake_news_dataset.csv")
```

```python
news_data.head()
```

```python
news_data.shape
```

```python
news_data.isnull().sum()
```

```python
#replace missing values with null string (NaN)

news_data = news_data.fillna("")
```

```python
#merge the author name and the title

news_data["content"] = news_data["author"] +" " +news_data["title"]
```

```python
news_data.head()
```

```python
X = news_data.drop(columns ="label", axis = 1)
Y = news_data["label"]
```

```python
print(X)
print(Y)
```

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
  if isinstance(content, list):  # Check if content is a list
    content = ' '.join(content)  # Join list elements into a string

  stemmed_content = re.sub("[^a-zA-Z]"," ",content)
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split()
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words("english")]
  stemmed_content = ' '.join(stemmed_content)
  return stemmed_content
```

```python
news_data['content'] = news_data['content'].apply(stemming)
```

```python
print(news_data["content"])
```

```python
X = news_data["content"]
Y = news_data["label"]
```

```python
print(X)
```

```python
print(Y)
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(X)

X = vectorizer.transform(X)
```

```python
print(X)
```
