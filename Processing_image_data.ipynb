Resizing the images using pillow

```python
from PIL import Image
```

```python
img = Image.open("/content/puppy image.jpeg")
```

```python
img_resize = img.resize((200,200))
```

```python
img_resize.save("dog image resized.jpg")
```

```python
img = mping.imread("/content/dog image resized.jpg")
```

```python
img.shape
```

```python
img_plot = plt.imshow(img)
img_plot
```

Practice on image processing

```python
import cv2
```

```python
img = cv2.imread("/content/puppy image.jpeg")
```
